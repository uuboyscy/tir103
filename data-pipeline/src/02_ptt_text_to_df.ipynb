```python
import os
import glob
# from pathlib import Path
import pandas as pd
```

```python
glob.glob("/workspaces/tir103/data-pipeline/res_gossiping/*.txt")
```

```python
columns = [
    "推",
    "噓",
    "分數",
    "作者",
    "標題",
    "時間",
]
data = []
for path in glob.glob("/workspaces/tir103/data-pipeline/res_gossiping/*.txt"):
    with open(path, "r") as f:
        ptt_file_string = f.read()

    some_string = ptt_file_string.split("---split---")[1]
    some_string_list = some_string.split("\n")[1:-1]

    # print(some_string_list)
    data.append(some_string_list)

data
```

```python
df = pd.DataFrame(
    data=data,
    columns=columns,
)

df
```

```python
df.to_csv("ptt.csv", index=False)
```

```python
df.to_csv("ptt_no_columns.csv", index=False, header=None)
```
